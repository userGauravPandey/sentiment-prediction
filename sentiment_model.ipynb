```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.txt.zip',sep=';',header=None,names=['text','emotion'])
```

```python
df['text']=df['text'].apply(lambda x:x.lower())
```

```python
import string
def remove_punc(txt):
    return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['text']=df['text'].apply(remove_punc)
```

```python
df.loc[2]['text']
```

```python
def remove_numbers(txt):
    new=''
    for i in txt:
        if not i.isdigit():
            new=new+i
    return new
df['text']=df['text'].apply(remove_numbers)
```

```python
df.loc[2]['text']
```

```python
df.loc[3]['text']
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words('english'))
```

```python
stop_words
```

```python
df.loc[2]['text']
```

```python
def remove(txt):
    words=txt.split()
    cleaned=[]
    for i in words:
        if not i in stop_words:
            cleaned.append(i)
    return ' '.join(cleaned)
df['text']=df['text'].apply(remove)
```

```python
df.loc[1]['text']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df['text'],df['emotion'],
test_size=0.20,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

```python
bow=CountVectorizer()
```

```python
x_train_b=bow.fit_transform(x_train)
x_test_b=bow.transform(x_test)
```

```python
x_train_b
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
nb_model=MultinomialNB()
```

```python
nb_model.fit(x_train_b,y_train)
```

```python
pred_nb=nb_model.predict(x_test_b)
```

```python
print(accuracy_score(y_test,pred_nb))
```

```python
y_test
```

```python
tf=TfidfVectorizer()
```

```python
x_train_tf=tf.fit_transform(x_train)
x_test_tf=tf.transform(x_test)
```

```python
nb2= MultinomialNB()
nb2.fit(x_train_tf,y_train)
```

```python
y_pred=nb2.predict(x_test_tf)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logi=LogisticRegression(max_iter=100)
```

```python
logi.fit(x_train_tf,y_train)
```

```python
log_pred=logi.predict(x_test_tf)
print(accuracy_score(y_test,log_pred))
```
